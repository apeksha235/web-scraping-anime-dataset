```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import re
```

```python
df=pd.read_csv(r'C:\Users\Apeksha\Documents\data_proj\anime_dataset.csv')
```

```python
df.head()
```

## Anime TV shows under Top 100

```python
tv_shows=df[df['No. of episodes']!=1]
```

```python
tv_shows
```

- ### **Graph for Rating vs No. of episodes**

```python
plt.plot(tv_shows['Rating'][:50],tv_shows['No. of episodes'][:50])
fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.xlabel("Rating")
plt.ylabel("No. of Episodes")
```

- ### **Rating for every Anime tv shows containing genre of 'Science fiction'**

```python
com_gen=tv_shows[tv_shows['Genre and Theme'].str.contains('science fiction')==True]
```

```python
plt.plot(com_gen['Name'],com_gen['Rating'])
fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Anime TV shows with 'Science fiction' genre ")
plt.ylabel("Rating")
```

- ### **Rating for Anime tv shows with genre 'Mystery'**

```python
my_gen=tv_shows[tv_shows['Genre and Theme'].str.contains('mystery')==True]
```

```python

x = my_gen.groupby('Name')['Rating'].mean().sort_values().tail(15)
x.plot(kind='barh')
plt.xlabel('Rating')
plt.ylabel('Anime TV shows')
```
