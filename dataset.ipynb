```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
import re
```

```python
url='https://www.animenewsnetwork.com/encyclopedia/ratings-anime.php?top50=best_bayesian&n=100'
response=requests.get(url)
```

```python
contents=response.text
with open('top-100-anime.html', 'w', encoding="utf-8") as file:
    file.write(contents)
```

Reading the HTML file of the website 

```python

with open('top-100-anime.html', 'r',encoding="utf-8") as f:
    html_source = f.read()
```

```python
doc = BeautifulSoup(html_source, 'html.parser')
```

Finding all the names of the anime under class 't'

```python
all_anime_raw=doc.find_all(class_='t')
```

```python
all_anime=all_anime_raw[1:]
```

Extracting the topic links 

```python
all_anime_links=doc.find(id='content-zone')
all_anime_links=all_anime_links('a')
all_anime_links=all_anime_links[8:108]
all_anime_links
```

```python
anime=[]
for i in all_anime_links:
 ex=re.search('>(.*)</a>',str(i))
 anime.append(ex.group(1))
anime=anime[:100]
```

```python
links=[]
base_url='https://www.animenewsnetwork.com/'
for i in range(0,100):
    links.append(base_url + all_anime_links[i]['href'])
links
```

Creating a function to extract from every topic url 

```python
def get_anime(u):
    response1=requests.get(u)
    contents1=response1.text
    with open('anime.html', 'w', encoding="utf-8") as file:
     file.write(contents1)
    with open('anime.html','r',encoding="utf-8") as f:
     html_sources = f.read()
    doc2 = BeautifulSoup(html_sources, 'html.parser')
    return doc2
    
```

Function to get the genre and theme for every anime

```python
def get_genre(soup):
    sub='/encyclopedia/search/genreresults'
    lists=[]
    genre1=soup.find_all(class_='discreet')
    for i in genre1:
        if(sub in str(i)):
          lists.append(str(i))
    genre=''
    for i in lists:
      genre=genre+(re.search('>(.*)<',str(i)).group(1))+","
    
    return genre
```

Creating a list of all the genres

```python
genre=[]
for i in links:
  doc1=get_anime(i)
  genre.append(get_genre(doc1))
genre
```

Function to get the number of episodes

```python
def get_ep(soup):
    get=soup.find_all('div',id='infotype-3')
    get=str(get)
    ep1=re.search('span>(.*)<',get)
    n_ep=ep1.group(1)
    n_ep=int(n_ep)   
    return n_ep
```

Creating a list of the number of episodes for tv shows, blank if it's a movie

```python
episodes=[]
for i in links:
  doc1=get_anime(i)
  try:
   episodes.append(get_ep(doc1))
  except AttributeError:
    episodes.append('')
episodes
```

```python
soup = BeautifulSoup(html_source, 'html.parser')
```

Extracting the rating of every anime

```python
star=soup.find_all('td',class_='r')
rating=[]
for i in star:
  exe=re.search('"r">(.*)<',str(i))
  rating.append(exe.group(1))
rating=rating[2::2]
rating
```

```python
import numpy as np
index=np.arange(1,101).tolist()
```

Creating a dataframe from the lists above

```python
dict={'Serial No.':index,'Name':anime,'Links':links,'Genre and Theme':genre,'No. of episodes':episodes,'Rating':rating}
df = pd.DataFrame(dict)
```
